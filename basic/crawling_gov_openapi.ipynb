# 공공데이터 OpenAPI 사용하기

## 0. 라이브러리 import

```python
import requests
from bs4 import BeautifulSoup
```

## 1. 기상 예측 정보 크롤링

```python
open_api = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109'

res = requests.get(open_api)
soup = BeautifulSoup(res.content, 'xml')
locations = soup.find_all('location')
```

```python
for location in locations:
    print(location.find('city').get_text())
    datas = soup.find_all('data')
    for data in datas:
        print(data.find('tmEf').get_text())
        print(data.find('wf').get_text())        
```

## 2. 아파트매매 실거래가 크롤링

```python
service_key = '' # put your service key
open_api = 'http://openapi.molit.go.kr:8081/OpenAPI_ToolInstallPackage/service/rest/RTMSOBJSvc/getRTMSDataSvcAptTrade'
location_code = '' # put your location code
deal_date = '202007'

request_api = open_api + '?LAWD_CD=' + location_code + '&DEAL_YMD=' + deal_date + '&serviceKey=' + service_key

res = requests.get(request_api)
soup = BeautifulSoup(res.content, 'xml')
```

```python
items = soup.find_all('item')

for item in items:
    print(item.find('아파트').get_text(), item.find('거래금액').get_text(), item.find('전용면적').get_text())
```
