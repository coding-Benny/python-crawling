# 네이버 Open API 사용하기

## 0. 라이브러리 import 및 client 정보 입력

```python
import requests
import pprint
```

```python
client_id = '' # put your client id
client_secret = '' # put your client secret
```

## 1. 특정 키워드 관련 뉴스 1000개에 해당하는 제목과 링크 가져오기

### 1-1. API 설정

```python
naver_open_api = 'https://openapi.naver.com/v1/search/news.json?query='
header_params = {'X-Naver-Client-Id': client_id, 'X-Naver-Client-Secret': client_secret}
query = '코로나'
options = '&display=100&start='
start_num = 1
```

### 1-2. 크롤링

```python
for index in range(10):
    start_num = 1 + (100 * index)
    res = requests.get(naver_open_api + query + options + str(start_num), headers = header_params)
    
    if res.status_code == 200:
        news_title = res.json()
        for title in news_title['items']:
            print(title['title'], title['originallink'])
```

## 2. 특정 키워드 관련 영화 제목, 제작년도, 평점, 링크 가져오기

### 2-1. API 설정

```python
naver_open_api = 'https://openapi.naver.com/v1/search/movie.json?query='
header_params = {'X-Naver-Client-Id': client_id, 'X-Naver-Client-Secret': client_secret}
query = '해리포터'
res = requests.get(naver_open_api + query, headers = header_params)
```

### 2-2. 크롤링

```python
if res.status_code == 200:
        movie_title = res.json()
        for movie in movie_title['items']:
            print(movie['title'], movie['pubDate'],  movie['userRating'], movie['link'])
```

## 3. 특정 키워드 관련 상품 리스트 가져오기

### 3-1. API 설정

```python
naver_open_api = 'https://openapi.naver.com/v1/search/shop.json?query='
header_params = {'X-Naver-Client-Id': client_id, 'X-Naver-Client-Secret': client_secret}
query = 'KF94 마스크'
options = '&display=100&sort=date&start='
start_num = 1
```

### 3-2. 크롤링

```python
product_type = {
    '1': "가격비교 일반상품",
    '2': "가격비교 비매칭 일반상품",
    '3': "가격비교 매칭 일반상품",
    '4': "가격비교 중고상품",
    '5': "가격비교 비매칭 중고상품",
    '6': "가격비교 매칭 중고상품",
    '7': "가격비교 단종상품",
    '8': "가격비교 비매칭 단종상품",
    '9': "가격비교 매칭 단종상품",
    '10': "가격비교 판매예정상품",
    '11': "가격비교 비매칭 판매예정상품",
    '12': "가격비교 매칭 판매예정상품"
}

for index in range(10):
    start_num = 1 + (100 * index)
    res = requests.get(naver_open_api + query + options + str(start_num), headers = header_params)
    
    if res.status_code == 200:
        product = res.json()
        for item in product['items']:
            print(item['title'], item['link'], item['lprice'], product_type[item['productType']])
```
